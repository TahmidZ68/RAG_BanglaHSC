```python
from pdf2image import convert_from_path
import pytesseract
import os

def extract_bangla_from_scanned_pdf(pdf_path):
    images = convert_from_path(pdf_path)
    bangla_text = ""
    for img in images:
        text = pytesseract.image_to_string(img, lang='ben')  # Bengali OCR
        bangla_text += text + "\n"
    return bangla_text.strip()

pdf_path = "/content/drive/MyDrive/Colab Notebooks/HSC26-Bangla1st-Paper.pdf"
raw_text = extract_bangla_from_scanned_pdf(pdf_path)
print(raw_text[:500])
```

```python
import re

def split_text_into_chunks(text, max_length=300):
    sentences = re.split(r'[।\n]', text)  # Bangla full stop or newline
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        sentence = sentence.strip()
        if not sentence:
            continue
        if len(current_chunk) + len(sentence) < max_length:
            current_chunk += sentence + "। "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + "। "
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

text_chunks = split_text_into_chunks(raw_text)
print(f"Total chunks: {len(text_chunks)}")
print(text_chunks[:3])  # Preview first 3 chunks
```

```python
import re

def chunk_text(text, max_len=200):
    # Basic sentence splitting using Bangla punctuation
    sentences = re.split(r'[।!?]', text)
    chunks = []
    current_chunk = ""

    for sentence in sentences:
        if len(current_chunk) + len(sentence) < max_len:
            current_chunk += sentence + "। "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + "। "

    if current_chunk:
        chunks.append(current_chunk.strip())

    return chunks

chunks = chunk_text(text)
print(f"Total Chunks: {len(chunks)}")
print("Sample Chunk:", chunks[0])
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load multilingual model
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')

# Convert text chunks to embeddings
embeddings = model.encode(text_chunks, show_progress_bar=True)

# Create FAISS index
dimension = embeddings[0].shape[0]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
def ask_question(query, top_k=3):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding), top_k)
    results = [text_chunks[i] for i in indices[0]]
    return results

# Example Bangla question
question = "তার মামা শম্ভুনাথ কে ছিল?"
answers = ask_question(question)

print("🧠 Top matching answers:")
for ans in answers:
    print("👉", ans)
```
